授權google sheet

```python
#授權
from google.colab import auth
auth.authenticate_user()

import gspread
from google.auth import default
creds, _ = default()

gc = gspread.authorize(creds)
```

```python
import pandas as pd
# read data and put it in a dataframe
# 在 google 工作表載入 gsheets
gsheets = gc.open_by_url('https://docs.google.com/spreadsheets/d/1Q_WjdN6BUN3Fet92pqQJhcu3wA2Po6xA9QOdb2CxYNo/edit?usp=sharing')
```

```python
from gspread_dataframe import set_with_dataframe

# 選擇要更新的工作表（選擇第2個工作表）
worksheet = gsheets.get_worksheet(1)  # 如果是其他工作表，可以用 get_worksheet(index)
worksheet
```

```python
# 取得 Google Sheets 內容
data = worksheet.get_all_values()  # 取得所有儲存格資料，回傳為 2D 陣列
df = pd.DataFrame(data)
df
```

AI內容

```python
from google.colab import userdata
apikey = userdata.get('GOOGLE_API_KEY')
```

```python
prompt=f"摘要{df.to_string()}並用繁體中文分析此公司狀況"
```

```python
import google.generativeai as genai
genai.configure(api_key = apikey)
model = genai.GenerativeModel("gemini-1.5-flash")
response = model.generate_content(prompt)
print(response.text)
```

```python
type(response.text)
```

```python
from io import StringIO

# 轉換為 DataFrame
df_all = pd.read_csv(StringIO(response.text))
df_all
```

```python
# 選擇要更新的工作表（選擇第3個工作表）
worksheet_allcommand = gsheets.get_worksheet(2)  # 如果是其他工作表，可以用 get_worksheet(index)
# 將 DataFrame 寫入 Google Sheets
set_with_dataframe(worksheet_allcommand, df_all)
```
