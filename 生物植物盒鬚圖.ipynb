figure.1.所有組別的10-5 M 面積增長(第五天面積減去第一天面積)分布


```python
import plotly.graph_objects as go
import numpy as np

# 數值處理
data_5["面積增長cm^2"] = data_5["面積增長cm^2"].astype(float).round(3)
values = data_5["面積增長cm^2"]

# 計算統計值
q1 = np.percentile(values, 25)
median = np.median(values)
q3 = np.percentile(values, 75)
iqr = q3 - q1
lower_fence = q1 - 1.5 * iqr
upper_fence = q3 + 1.5 * iqr

# 在 fence 範圍內找 min, max
fence_values = values[(values >= lower_fence) & (values <= upper_fence)]
min_val = fence_values.min()
max_val = fence_values.max()

# 建立圖
fig = go.Figure()

# 畫盒鬚圖
fig.add_trace(go.Box(
    y=values,
    name='10^-5 M',
    boxpoints='all',
    jitter=0.4,
    pointpos=0,
    marker=dict(color='lightblue'),
    line=dict(color='blue')
))

# 疊加統計文字標籤
stats = [
    ("Q1", q1),
    ("Median", median),
    ("Q3", q3),
    ("IQR", iqr),
    ("Lower Fence", lower_fence),
    ("Upper Fence", upper_fence),
    ("Min", min_val),
    ("Max", max_val)
]

for label, y in stats:
    fig.add_trace(go.Scatter(
        x=['10^-5 M'],
        y=[y],
        text=[f"{label}: {round(y, 3)}"],
        mode='text',
        textposition="middle right",  # 你也可以改為 "top center"
        textfont=dict(size=10, color="black"),
        showlegend=False
    ))

# 美化圖表
fig.update_layout(
    title="10^-5 M 面積增長盒鬚圖",
    yaxis_title="面積增長 cm²"
)

fig.show()
```

```python
import plotly.graph_objects as go
import numpy as np

# 數值處理
df_GA["面積增長cm^2"] = df_GA["面積增長cm^2"].astype(float)

fig = go.Figure()

# 對每個濃度分組處理
for conc in df_GA["濃度"].unique():
    subset = df_GA[df_GA["濃度"] == conc]["面積增長cm^2"]
    values = subset.values

    # 計算統計值
    q1 = np.percentile(values, 25)
    median = np.median(values)
    q3 = np.percentile(values, 75)
    iqr = q3 - q1
    lower_fence = q1 - 1.5 * iqr
    upper_fence = q3 + 1.5 * iqr
    fence_values = values[(values >= lower_fence) & (values <= upper_fence)]
    min_val = fence_values.min()
    max_val = fence_values.max()

    # 畫箱型圖
    fig.add_trace(go.Box(
        y=values,
        x=[conc] * len(values),
        name=conc,
        boxpoints='all',
        jitter=0.4,
        pointpos=0,
        marker=dict(color='lightgreen'),
        line=dict(color='green'),
        showlegend=False
    ))

    # 加上統計數字
    stats = [
        ("Q1", q1),
        ("Median", median),
        ("Q3", q3),
        ("IQR", iqr),
        ("Lower Fence", lower_fence),
        ("Upper Fence", upper_fence),
        ("Min", min_val),
        ("Max", max_val)
    ]

    for label, y in stats:
        fig.add_trace(go.Scatter(
            x=[conc],
            y=[y],
            text=[f"{label}: {round(y, 3)}"],
            mode='text',
            textposition="middle right",
            textfont=dict(size=13, color="black"),
            showlegend=False
        ))

# 更新圖表
fig.update_layout(
    title="不同濃度GA的面積增長 cm² 分布（含統計數值）",
    xaxis_title="濃度 (M)",
    yaxis_title="面積增長 cm²"
)

fig.show()
```

figure.4.比較GA和BA在分別濃度下子葉重量增長平均

figure.5.比較GA和BA在分別濃度下子葉面積增長平均

```python
import pandas as pd
import plotly.graph_objects as go

# 數值轉型
for df in [df_GA, df_BA, df_ori]:
    df["面積增長cm^2"] = pd.to_numeric(df["面積增長cm^2"], errors="coerce")
    df["重量增長g"] = pd.to_numeric(df["重量增長g"], errors="coerce")

# 加入藥物標籤
df_GA["藥物"] = "GA"
df_BA["藥物"] = "BA"
df_ori["藥物"] = "對照組"

# 合併原始資料，供群組統計用
df_all = pd.concat([df_GA, df_BA, df_ori])

# 群組統計
grouped_df = df_all.groupby(["藥物", "濃度"]).agg({
    "面積增長cm^2": ["mean", "std"],
    "重量增長g": ["mean", "std"]
}).reset_index()

# 重新命名欄位
grouped_df.columns = ["藥物", "濃度", "面積平均", "面積標準差", "重量平均", "重量標準差"]

# 設定濃度順序
grouped_df["濃度"] = pd.Categorical(grouped_df["濃度"],
    categories=["對照組","10^-4 M", "10^-5 M", "10^-6 M", "10^-7 M"], ordered=True)

grouped_df = grouped_df.sort_values(by=["藥物", "濃度"])

# === 繪製 面積圖 ===
fig_area = go.Figure()

for drug in grouped_df["藥物"].unique():
    sub_df = grouped_df[grouped_df["藥物"] == drug]
    fig_area.add_trace(go.Bar(
        name=drug,
        x=sub_df["濃度"],
        y=sub_df["面積平均"],
        error_y=dict(type='data', array=sub_df["面積標準差"], visible=True),
        text=sub_df["面積平均"].round(2),
        textposition='auto',
        textfont=dict(size=12)
    ))

fig_area.update_layout(
    title="GA、BA 與對照組對子葉面積增長的影響",
    xaxis_title="濃度",
    yaxis_title="子葉面積增長 (cm²)",
    barmode='group',
    template='plotly_white',
    xaxis=dict(
        categoryorder='array',
        categoryarray=["對照組","10^-4 M", "10^-5 M", "10^-6 M", "10^-7 M"]
    )
)

fig_area.show()

# === 繪製 重量圖 ===
fig_weight = go.Figure()

for drug in grouped_df["藥物"].unique():
    sub_df = grouped_df[grouped_df["藥物"] == drug]
    fig_weight.add_trace(go.Bar(
        name=drug,
        x=sub_df["濃度"],
        y=sub_df["重量平均"],
        error_y=dict(type='data', array=sub_df["重量標準差"], visible=True),
        text=sub_df["重量平均"].round(3),
        textposition='auto',
        textfont=dict(size=12)
    ))

fig_weight.update_layout(
    title="GA、BA 與對照組對子葉重量增長的影響",
    xaxis_title="濃度",
    yaxis_title="子葉重量增長 (g)",
    barmode='group',
    template='plotly_white',
    xaxis=dict(
        categoryorder='array',
        categoryarray=["對照組","10^-4 M", "10^-5 M", "10^-6 M", "10^-7 M"]
    )
)

fig_weight.show()

```

圖表比較與解釋(GA 與 BA 不同濃度的效果比較)

Figure 2 與 Figure 3：面積增長分布

Figure 2 與 Figure 3 分別呈現 GA 與 BA 在不同濃度下子葉面積的增長（第5天面積減去第0天面積）分布情況。對於 GA（吉貝素）而言，其在中高濃度（10⁻⁴ M 與 10⁻⁵ M）下展現穩定且明顯的促進效果，顯示 GA 在高濃度下能有效刺激葉面積增長。這種作用可能來自其促進細胞分裂與細胞伸長的雙重生理機制。相比之下，BA（細胞分裂素）在 10⁻⁶ M 濃度時增長效果最佳，而在 10⁻⁷ M 或 10⁻⁵ M 的表現則較不穩定。此現象顯示 BA 的效果對濃度較為敏感，可能存在一個較窄的最佳濃度範圍，高於或低於該範圍皆可能抑制其效果。

Figure 4：重量增長比較

在重量增長部分，GA 在 10⁻⁴ M 與 10⁻⁵ M 表現接近，其中 10⁻⁴ M 略勝一籌。相較之下，BA 於 10⁻⁶ M 表現最佳，但整體重量增長仍不及 GA 的中高濃度表現。
另外值得注意的是，BA 在 10⁻⁵ M 與 10⁻⁷ M 兩個濃度下增長程度相近，並未呈現明顯的濃度依賴性，顯示其在重量促進方面同樣具有濃度敏感性。兩者相比於對照組（純營養劑）均有較高表現，驗證生長劑的正向促進作用。

Figure 5：面積增長平均比較

面積增長結果顯示，GA 在中高濃度下穩定提升葉片面積，表現優於 BA，且與重量增長趨勢一致。不過，BA 在 10⁻⁵ M 的面積增長略高於 10⁻⁷ M，顯示其對面積的影響可能相較於重量更明顯。

整體而言，GA 無論在重量或面積的提升上皆表現穩定而顯著；BA 則展現一定的促進作用，但效果對濃度較為敏感且不穩定。
