```python
import plotly.express as px
import pandas as pd

# 確保數據型態為數值
data_who["離婚人數中大陸港澳及其他外籍配偶之人數比率[%]"] = pd.to_numeric(data_who["離婚人數中大陸港澳及其他外籍配偶之人數比率[%]"], errors="coerce")

# 繪製折線圖
fig = px.line(
    data_who,
    x="年別",  # X 軸：年份
    y="離婚人數中大陸港澳及其他外籍配偶之人數比率[%]",  # Y 軸：比率
    title="離婚人數中大陸港澳及其他外籍配偶之人數比率",  # 標題
    labels={"離婚人數中大陸港澳及其他外籍配偶之人數比率[%]": "比率 (%)", "年別": "年份"}  # 標籤
)

# 顯示圖表
fig.show()
```

```python
import plotly.express as px
import pandas as pd
data_bgrate = pd.DataFrame(data_date[12:])
# 確保數據型態為數值
data_bgrate["離婚平均年齡/男[歲]"] = pd.to_numeric(data_date["離婚平均年齡/男[歲]"], errors="coerce")
data_bgrate["離婚平均年齡/女[歲]"] = pd.to_numeric(data_date["離婚平均年齡/女[歲]"], errors="coerce")

# 繪製雙折線圖
fig = px.line(
    data_bgrate,
    x="年別",  # X 軸：年份
    y=["離婚平均年齡/男[歲]", "離婚平均年齡/女[歲]"],  # Y 軸：男、女的離婚平均年齡
    title="離婚平均年齡 / 男性與女性",  # 標題
    labels={"年別": "年份",
            "離婚平均年齡/男[歲]": "男性平均年齡",
            "離婚平均年齡/女[歲]": "女性平均年齡"}  # 標籤
)

# 顯示圖表
fig.show()
```
